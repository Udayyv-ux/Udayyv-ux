```python
import warnings
warnings.filterwarnings('ignore')
import zipfile
import pandas as pd
import urllib.request
import io
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00360/AirQualityUCI.zip'
response = urllib.request.urlopen(url)
```

```python
zip_file = zipfile.ZipFile(io.BytesIO(response.read()))
with zip_file.open('AirQualityUCI.csv') as file:
    data = pd.read_csv(file, sep=';', decimal=',', parse_dates={'Datetime': ['Date', 'Time']}, na_values=-200)
data.head()
```

```python
data_pd = data.copy()
data_pd.head()
```

```python
data_pd.isnull().sum()
```

```python
data_pd = data_pd.dropna(axis=1,how='all')
data_pd.head()
```

```python
data_pd = data_pd.dropna()
data_pd
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt

data_pd = data_pd.drop(columns=['Datetime'])

data_pd = data_pd.apply(pd.to_numeric, errors='coerce')

data_pd = data_pd.dropna(subset=['C6H6(GT)'])

X = data_pd.drop(columns=['C6H6(GT)'])
y = data_pd['C6H6(GT)']

X_log_transformed = np.log1p(X)

X_boxcox_transformed = X_log_transformed + abs(X_log_transformed.min()) + 1
pt = PowerTransformer(method='box-cox')
X_boxcox_transformed = pt.fit_transform(X_boxcox_transformed)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_boxcox_transformed)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

y_pred = lr_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2:.4f}')
print(f'Mean Squared Error: {mse:.4f}')

new_features = X_test[:1]
new_prediction = lr_model.predict(new_features)
print(f'Prediction for new set of features: {new_prediction}')
```
