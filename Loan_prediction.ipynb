```python
lr_data = data_lp.copy()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr  = LogisticRegression()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
columns = lr_data.select_dtypes(include='object').columns
```

```python
columns
```

```python
for i in columns:
    lr_data[i] = le.fit_transform(lr_data[i])
```

```python
lr_data.head()
```

```python
for i in range(lr_data['Credit_Worthiness'].max()):
    print(f'{i} => {le.inverse_transform([i])}')
```

```python
X = lr_data.drop(columns=['Credit_Worthiness'])
```

```python
Y = lr_data['Credit_Worthiness']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
ss = StandardScaler()
xtrain_ss = ss.fit_transform(xtrain)
xtest_ss = ss.transform(xtest)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

lr.fit(xtrain_ss,ytrain)
```

```python
ypred_lr = lr.predict(xtest_ss)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score,f1_score, confusion_matrix
```

```python
acc_lr = accuracy_score(ytest,ypred_lr)
```
